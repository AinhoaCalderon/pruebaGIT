```python
# Leer el archivo PDF y devuelve las páginas que no son vacaciones
def getRNTNormalPages(pdfName):
  # we define the box containing margins. Margins must be expressed in pdf points. 
  # However, our PDF visualizer gives them in mm. Thus, we transform them in cm and then we convert them in 
  # pdf points by multiplying them for a conversion factor.

  # [top,left,bottom,width] - Cabecera coordinates
  box = [4.11,1.24,6.54,10.95]
  cf = 28.28
  for i in range(0, len(box)):
    box[i] *= cf

  
  # Extraer tablas del PDF
  tablesDF = tabula.read_pdf(pdfName, area=[box], pages='all', multiple_tables=True, silent=True, stream=True) # silent=True evita que aparezcan warnings
  #print(tablesDF)
  page=1
  paginas=[]
  for tableDF in tablesDF:
    tableDF.rename(columns={ tableDF.columns[0]: "Concepto" , tableDF.columns[1]: "Valor"}, inplace = True)
    tableDF['Page']=page
    page=page+1
    
    tabla_filtrada=tableDF[(tableDF['Concepto']=='Calificador de la liquidación') & (tableDF['Valor']=='L00-NORMAL')]
    #print(tabla_filtrada.size)
    if(tabla_filtrada.size>0):
      paginas.append(tabla_filtrada.iloc[0,2])
      #print(tabla_filtrada.iloc[0,2])
  return (paginas)


#print('Páginas ', getRNTNormalPages('Reto_RNT_Febrero2023.pdf'))

def getTablesDF(pdfName,paginasP):
  pdfRNT=tabula.read_pdf(pdfName, encoding="ISO-8859-1", 
              stream=True, area=(196.5,34.6,815.5,578), pages=paginasP, guess=False, 
              pandas_options={"header":None})
  
  return(pdfRNT)

def cleanColumns(tableDF):
  return(tableDF)

def cleanRows(tableDF):
  #filtrar filas innecesarias -null, NaN,...-

  #limpiar suma final de cada página
  return(tableDF)

def cleanTable(tableDF):
  tableDF=cleanColumns(tableDF)
  tableDF=cleanRows(tableDF)
  return(tableDF)

def autocompleteDNICells(tableDF):
  
  for i in range(0, len(tableDF)):
     if not tableDF[i]:
         tableDF[i] = tableDF[i-1]
  return(tableDF)

def autocompleteDateTimeCells(tableDF):
  return(tableDF)

def autocompleteCells(tableDF):
  #tableDF['DNI']=pd.Series(autocompleteDNICells(tableDF))
  tableDF=autocompleteDateTimeCells(tableDF)
  return(tableDF)

def getFinalRowsDF(tableDF):
  return(tableDF)

def getFirstSheetDF(tableDF):
  return(tableDF)

def getSecondSheetDF(tableDF):
  return(tableDF)

def writeExcel(firstSheetDF,secondSheetDF):
  return()
```
